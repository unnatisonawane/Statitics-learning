## Bernoulli Distribution

Bernoulli Distribution:

1. models a random experiment with two possible outcomes: success (usually denoted as 1) or failure (usually denoted as 0).

2. PMF is given by:

P(X = k) = p^k * (1 - p)^(1 - k) where X is the random variable, k is the outcome (0 or 1), and p is the probability of success. Example: Tossing a fair coin. Assume P(H) = 0.5 and P(T) = 0.5

```python
import numpy as np
from scipy.stats import bernoulli

# Define the probability of success
p = 0.7

# Create a Bernoulli distribution object
dist = bernoulli(p)

# Generate a random sample of size 10 from the distribution
sample = dist.rvs(size=10)

# Print the sample
print("Sample:", sample)

# Calculate the probability mass function (PMF) at a specific value
x = 1
pmf = dist.pmf(x)
print("PMF at x={}: {:.3f}".format(x, pmf))

# Calculate the cumulative distribution function (CDF) at a specific value
x = 0
cdf = dist.cdf(x)
print("CDF at x={}: {:.3f}".format(x, cdf))

# Calculate the mean and variance of the distribution
mean = dist.mean()
variance = dist.var()
print("Mean:", mean)
print("Variance:", variance)
```

## Binomial Distribution

1. models the number of successes in a fixed number of independent Bernoulli trials.
2. The PMF formula for the binomial distribution is:

PMF(k) = C(n, k) * p^k * (1 - p)^(n - k)
Where:

C(n, k) is the binomial coefficient, calculated as n! / (k! * (n - k)!)

p is the probability of success in each trial

k is the number of successes

n is the number of trials

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# Define the number of trials and the probability of success
n = 10
p = 0.5

# Create a binomial distribution object
dist = binom(n, p)

# Generate a random sample of size 10 from the distribution
sample = dist.rvs(size=10)

# Print the sample
print("Sample:", sample)

# Calculate the probability mass function (PMF) at a specific value
x = 5
pmf = dist.pmf(x)
print("PMF at x={}: {:.3f}".format(x, pmf))

# Calculate the cumulative distribution function (CDF) at a specific value
x = 7
cdf = dist.cdf(x)
print("CDF at x={}: {:.3f}".format(x, cdf))

# Calculate the mean and variance of the distribution
mean = dist.mean()
variance = dist.var()
print("Mean:", mean)
print("Variance:", variance)

# Plot the histogram
plt.hist(sample, bins=np.arange(n+2)-0.5, edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution Histogram')
plt.xticks(range(n+1))
plt.show()
```

## Poisson Distribution

models the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence.

```python
import numpy as np
from scipy.stats import poisson

# Define the average rate of occurrence
lambda_ = 2.5

# Create a Poisson distribution object
dist = poisson(mu=lambda_)

# Generate a random sample of size 10 from the distribution
sample = dist.rvs(size=10)

# Print the sample
print("Sample:", sample)

# Calculate the probability mass function (PMF) at a specific value
x = 3
pmf = dist.pmf(x)
print("PMF at x={}: {:.3f}".format(x, pmf))

# Calculate the cumulative distribution function (CDF) at a specific value
x = 5
cdf = dist.cdf(x)
print("CDF at x={}: {:.3f}".format(x, cdf))

# Calculate the mean and variance of the distribution
mean = dist.mean()
variance = dist.var()
print("Mean:", mean)
print("Variance:", variance)

plt.hist(sample, bins=np.arange(np.max(sample) + 2) - 0.5, edgecolor='black')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.title('Poisson Distribution Histogram')
plt.show()
```

## Normal Distribution

1. It is a continuous probability distribution that is symmetric and bell-shaped.
2. The mean determines the location of the peak of the distribution.
3. The standard deviation controls the width of the distribution.
4. A smaller standard deviation resulting in a narrower distribution. Conversely, a larger standard deviation leads to a wider distribution with more spread-out data points.

```python
import numpy as np
from scipy.stats import norm

# Define the mean and standard deviation
mean = 0
std = 1

# Create a normal distribution object
dist = norm(loc=mean, scale=std)

# Calculate the PDF at a specific value
x = 1.5
pdf = dist.pdf(x)
print("PDF at x={}: {:.4f}".format(x, pdf))

# Calculate the CDF at a specific value
x = 0.8
cdf = dist.cdf(x)
print("CDF at x={}: {:.4f}".format(x, cdf))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the mean and standard deviation
mean = 0
std = 1

# Generate a random sample from the normal distribution
sample_size = 1000
sample = np.random.normal(mean, std, size=sample_size)

# Plot the histogram
plt.hist(sample, bins=30, edgecolor='black', density=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution Histogram')
plt.show()
```

## Standard Normal Distribution

Standard Normal Distribution: The z-score is specifically used in reference to the standard normal distribution, which has a mean of 0 and a standard deviation of 1
1. The z-score, also known as the standard score, represents the number of standard deviations an observation or data point is away from the mean of a distribution.
2. The formula to calculate the z-score is:
z = (x - μ) / σ
where z is the z-score, x is the value, μ is the mean of the distribution, and σ is the standard deviation of the distribution.


```python
import numpy as np
import scipy.stats as stats

# Define the data point, mean, and standard deviation
x = 75
mean = 70
std = 8

# Calculate the z-score
z = (x - mean) / std

print("Z-score:", z)

# Calculate the probability using the survival function (1 - cumulative distribution function)
probability =stats.norm.cdf(z)

print("Probability:", probability)
```

## Central Limit Thm

The Central Limit Theorem (CLT) is a fundamental theorem in statistics that states that the distribution of the sample mean from a population with any distribution approaches a normal distribution as the sample size increases, regardless of the shape of the population distribution.

1. The CLT allows us to make inferences about the population mean based on the distribution of sample means
2. as the sample size increases, the sampling distribution of the mean becomes approximately normally distributed
3. even if the population is not normally distributed, as long as the sample size is sufficiently large, the distribution of the sample mean will still approximate a normal distribution.
